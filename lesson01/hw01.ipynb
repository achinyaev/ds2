### Домашнее задание №1 <br>
1. Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.<br>
2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?<br>

1. Примеры:<br>

Точность<br>
Прогноз погоды - для погоды важна точность предсказаний, например когда находишься на природе, на воде или в лесу, так как от этого может зависеть жизнь человека. По этому для людей котором важен прогноз погоды, следует передавать весь объем информации по возможному изменению погоды. <br>
При этом прогноз погоды может быть не важен когда сидишь дома. т.е. все зависит от  требований к предсказанию.<br>

Широта<br>
Диагностирование заболевания в период эпидемии — в такой период важно как можно больше произвести обследований населения и тем самым выявить как можно больше подходящих под симптомы заболевших людей, что бы отправить их на карантин и начать лечение. При этом не так важна точность выявления именно самого заболевания вызвавшего эпидемию, так как лечение симптоматичное и ведет только к устранению симптомов для облегчения протекания болезни, так как лекарства для лечения именного самого заболевания нет. <br>

2. F-мера <br>

F-мера позволяет определить на сколько влияет на нашу модель точность или полнота. Измения коэффициент бета от 0 и более 1, мы можем понять по графику F-меры какой коэф. больше оказывает влияние на метрику. и над чем нужно порабоать в нашей модели.
